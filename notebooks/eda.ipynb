### Conclusions after a first inspection (Regarding to data quality):

1. There are some missing values in all datasets but they are not a representative quantity in comparison to datasets size.

2. I can note that the datatypes in self-assessment there is a problematic register on several columns -> '#VALUES'. I'll change that for Nan an then I'll convert the datatypes to numeric. 

3. Some datatypes on superior-assessment have to be changed from string to numeric.

```python
# Making corrections

for column in auto.columns:
    auto[column] = auto[column].apply(lambda val: np.nan if val == '#VALUE!' else val)

for i in range(1,13):
    auto[auto.columns[i]] = auto[auto.columns[i]].astype('float')

for i in range(4,16):
    sup[sup.columns[i]] = sup[sup.columns[i]].astype('float')
```

```python
# Checking if corrections were correctly applied
sv.DataframeReport(auto).show_notebook()
```
